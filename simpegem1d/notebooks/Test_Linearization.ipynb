```python
from SimPEG import *
from simpegem1d import EM1D, EM1DAnal, Kernels, RTEfun, DigFilter, BaseEM1D, Waveform
from scipy.constants import mu_0
from scipy.interpolate import interp1d
%pylab inline
```

```python
TDsurvey = BaseEM1D.EM1DSurveyTD()
TDsurvey.rxLoc = np.array([0., 0., 100.+30.])
TDsurvey.txLoc = np.array([0., 0., 100.+30.])
TDsurvey.fieldtype = 'secondary'
TDsurvey.rxType = 'Bz'
TDsurvey.waveType = 'stepoff'
TDsurvey.txType = 'CircularLoop'
I = 1e0
a = 1e1
TDsurvey.I = I
TDsurvey.a = a

nearthick = np.logspace(-1, 1, 2)
deepthick = np.logspace(1, 2, 2)
hx = np.r_[nearthick, deepthick]
mesh1D = Mesh.TensorMesh([hx], [0.])
depth = -mesh1D.gridN
LocSigZ = np.r_[-mesh1D.gridCC, -mesh1D.gridCC[-1]-mesh1D.gridCC[-1]]
nlay = depth.size
topo = np.r_[0., 0., 100.]
TDsurvey.depth = depth
TDsurvey.topo = topo
TDsurvey.LocSigZ = LocSigZ
TDsurvey.HalfSwitch = True
TDsurvey.Setup1Dsystem()
TDsurvey.time = np.logspace(-14, -2.5, 1024)

TDsurvey.setFrequency(TDsurvey.time)
chi_half = 0.

Logmodel = BaseEM1D.BaseEM1DModel(mesh1D)
tau_blk = 1e-3
eta_blk = 9e-1
c_blk = 0.8
# blkind = (LocSigZ>-200) & (LocSigZ<-100)
blkind = 0
sig_half = 0.01 
sig_blk = 0.01
sig = np.ones(nlay)*sig_half
sig[blkind] = sig_blk
tau = np.zeros(nlay)
tau[blkind] = tau_blk
eta = np.zeros(nlay)
eta[blkind] = eta_blk
c = np.zeros(nlay)
c[blkind] = c_blk
# options = {'Frequency': TDsurvey.frequency, 'tau': np.ones(nlay)*tau_half, 'eta':np.ones(nlay)*eta_half, 'c':np.ones(nlay)*c_half}
options = {'Frequency': TDsurvey.frequency, 'tau': tau, 'eta':eta, 'c':c}
Colemodel = BaseEM1D.BaseColeColeModel(mesh1D, **options)

modelReal = Model.ComboModel(mesh1D, [Logmodel])
modelComplex = Model.ComboModel(mesh1D, [Colemodel, Logmodel])                
m_1D = np.log(np.ones(nlay)*sig)

WT0 = np.load('../WT0.npy')
WT1 = np.load('../WT1.npy')
YBASE = np.load('../YBASE.npy')
options = {'WT0': WT0, 'WT1': WT1, 'YBASE': YBASE}
# Question to rowan: can I share same survey object
```

```python
probEM = EM1D.EM1D(modelReal, **options)
if probEM.ispaired:
    probEM.unpair()
if TDsurvey.ispaired:
    TDsurvey.unpair()
probEM.pair(TDsurvey)
probEM.chi = np.zeros(TDsurvey.nlay)
probEM.CondType = 'Real'        
m_1D = np.log(sig)
probEM.jacSwitch = True
Hz, dHzdsig = probEM.fields(m_1D)
probEM.survey.rxType = 'Bz'
BzTD = probEM.survey.projectFields(u=Hz)
dBzdsigTD = probEM.survey.projectFields(u=dHzdsig)
probEM.survey.rxType = 'dBzdt'
dBzdtTD = probEM.survey.projectFields(u=Hz)
ddBzdtdsigTD = probEM.survey.projectFields(u=dHzdsig)
```

```python
dHzdsigr = Utils.mkvc(dHzdsig.real)
dHzdsigi = Utils.mkvc(dHzdsig.imag)

fig, axes = subplots(1,2, figsize = (14,6))
axes[0].loglog(probEM.survey.frequency[dHzdsigr>0], dHzdsigr[dHzdsigr>0], 'b', probEM.survey.frequency[dHzdsigr<0], -dHzdsigr[dHzdsigr<0], 'b--')
axes[1].loglog(probEM.survey.frequency[dHzdsigi>0], dHzdsigi[dHzdsigi>0], 'r', probEM.survey.frequency[dHzdsigi<0], -dHzdsigi[dHzdsigi<0], 'r--')
# axes[0].set_xlim(1e-5, 1e5);   axes[1].set_xlim(1e-5, 1e5)
# axes[0].set_ylim(1e-18, 1e0);  axes[1].set_ylim(1e-18, 1e0)
axes[0].set_xlabel('Frequency (Hz)', fontsize = 16)
axes[0].set_ylabel('Real [$dB_z / d\sigma_{\infty}$]', fontsize = 16)
axes[1].set_xlabel('Frequency (Hz)', fontsize = 16)
axes[1].set_ylabel('Imaginary [$dB_z / d\sigma_{\infty}$]', fontsize = 16)
axes[0].grid(color='k', alpha=0.5, linestyle='dashed', linewidth=0.5)
axes[1].grid(color='k', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

```python
probEMIP = EM1D.EM1D(modelComplex, **options)
if probEMIP.ispaired:
    probEMIP.unpair()
if TDsurvey.ispaired:
    TDsurvey.unpair()
probEMIP.pair(TDsurvey)
probEMIP.chi = np.zeros(TDsurvey.nlay)
probEMIP.CondType = 'Complex'        
m_1D = np.log(sig)
probEMIP.survey.rxType = 'Bz'
HzCole = probEMIP.fields(m_1D)
BzTDCole = probEMIP.survey.projectFields(u=HzCole)
probEMIP.survey.rxType = 'dBzdt'
dBzdtTDCole = probEMIP.survey.projectFields(u=HzCole)
```
