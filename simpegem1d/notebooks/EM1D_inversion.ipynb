```python
from SimPEG import *
from simpegem1d import EM1D, EM1DAnal, Kernels, RTEfun, DigFilter, BaseEM1D
%pylab inline
```

```python
FDsurvey = BaseEM1D.EM1DSurveyFD()
FDsurvey.rxLoc = np.array([0., 0., 100.+30.])
FDsurvey.txLoc = np.array([0., 0., 100.+30.])
FDsurvey.fieldtype = 'secondary'
FDsurvey.rxType = 'Hz'
FDsurvey.txType = 'VMD'
FDsurvey.offset = 8.
cs = 10.
nearthick = np.logspace(-1, 1, 3)
linthick = np.ones(10)*cs
deepthick = np.logspace(1, 2, 3)
hx = np.r_[nearthick, linthick, deepthick, deepthick[-1]]
mesh1D = Mesh.TensorMesh([hx], [0.])
depth = -mesh1D.gridN[:-1]
LocSigZ = -mesh1D.gridCC
nlay = depth.size
topo = np.r_[0., 0., 100.]
FDsurvey.depth = depth
FDsurvey.topo = topo
FDsurvey.LocSigZ = LocSigZ
# FDsurvey.frequency = np.logspace(3, 5, 11)
FDsurvey.frequency = np.r_[900., 7200., 56000]
FDsurvey.Nfreq = FDsurvey.frequency.size
FDsurvey.Setup1Dsystem()
sig_half = 1e-3
sig_blk = 1e-2
chi_half = 0.
Logmodel = BaseEM1D.BaseEM1DModel(mesh1D)
model = Model.ComboModel(mesh1D, [Logmodel])
```

```python
sig  = np.ones(nlay)*sig_half
blk_ind = (0>LocSigZ) & (-30<LocSigZ)
sig[blk_ind] = sig_blk
m_true = np.log(sig)
```

```python
fig, ax = subplots(1,1, figsize=(5, 8))
ax.plot(sig, LocSigZ,  'b.-')
ax.grid(color='k', alpha=0.5, linestyle='dashed', linewidth=0.5)
ax.set_xscale('log')
ax.set_xlim(1e-4, 1e1)
```

```python
WT0, WT1, YBASE = DigFilter.LoadWeights()
options = {'WT0': WT0, 'WT1': WT1, 'YBASE': YBASE}
prob = EM1D.EM1D(model, **options)
if prob.ispaired:
    prob.unpair()
if FDsurvey.ispaired:
    FDsurvey.unpair()
prob.pair(FDsurvey)
prob.CondType = 'Real'
prob.chi = np.zeros(FDsurvey.nlay)
d_true = FDsurvey.dpred(m_true)
```

```python
fig, axes = subplots(1,1, figsize = (7,5))
axes.plot(FDsurvey.frequency, -d_true[0:FDsurvey.Nfreq], '.-')
axes.plot(FDsurvey.frequency, -d_true[FDsurvey.Nfreq:], '.-')
axes.set_xscale('log')
```

```python
FDsurvey.dtrue = d_true
std = 0.05
noise = std*abs(FDsurvey.dtrue)*np.random.randn(*FDsurvey.dtrue.shape)
FDsurvey.dobs = FDsurvey.dtrue+noise
FDsurvey.std = FDsurvey.dobs*0 + std
FDsurvey.Wd = 1/(abs(FDsurvey.dobs)*std)
prob.jacSwitch = True
```

```python
opt = Optimization.InexactGaussNewton(maxIter = 10)
reg = Regularization.Tikhonov(model)
obj = ObjFunction.BaseObjFunction(FDsurvey, reg, beta=beta)
# Create an inversion object
inv = Inversion.BaseInversion(obj, opt)
m0 = np.log(np.ones(nlay)*sig_half)
reg.alpha_s = 1e-2
reg.alpha_x = 1.
C =  Utils.Counter()
prob.counter = C
opt.counter = C
opt.LSshorten = 0.5
opt.remember('xc')
beta = Parameters.BetaSchedule(coolingFactor = 8, coolingRate = 2, beta0_ratio = 1e2)
inv.objFunc.beta = beta
```

```python
mopt = inv.run(m0)
```

```python
opt.counter.summary()
xc = opt.recall('xc')
```

```python
dpredall = []
for i in range(opt.maxIter):
    dpredtemp = FDsurvey.dpred(xc[i])
    dpredall.append(dpredtemp)
```

```python
from JSAnimation import IPython_display
from matplotlib import animation

fig, ax = subplots(1,2, figsize=(14, 8))
ax[0].plot(sig, LocSigZ,  'b.-')
ax[0].grid(color='k', alpha=0.5, linestyle='dashed', linewidth=0.5)
ax[1].grid(color='k', alpha=0.5, linestyle='dashed', linewidth=0.5)
ax[1].plot(FDsurvey.frequency, -d_true[0:FDsurvey.Nfreq], 'b-')
ax[1].plot(FDsurvey.frequency, -d_true[FDsurvey.Nfreq:], 'r-')
ax[0].set_xscale('log')
ax[1].set_xscale('log')
ax[0].set_xlim(1e-5, 1e0)


frame1, = ax[0].plot([], [],  'r.-', lw=2)
frame2, = ax[1].plot([], [],  'b*' , lw=2)
frame3, = ax[1].plot([], [],  'r*' , lw=2)

def animate(i_id):
    
    frame1.set_data(model.transform(xc[i_id]), LocSigZ)
    frame2.set_data(FDsurvey.frequency, abs(dpredall[i_id][:FDsurvey.Nfreq]))
    frame3.set_data(FDsurvey.frequency, abs(dpredall[i_id][FDsurvey.Nfreq:]))
    
    ax[0].legend(('True', 'Pred'), loc=2, fontsize = 14)
    ax[1].legend(('$d^{real}_{Obs}$', '$d^{imag}_{Obs}$', '$d^{real}_{Pred}$', '$d^{imag}_{Pred}$'), loc=2, fontsize = 18)
    ax[0].set_title('$\sigma$ model, Iter = ' + str(i_id+1), fontsize = 16)
    ax[1].set_title('Pred and Obs data, Iter = ' + str(i_id+1), fontsize = 16)
    return frame1, frame2, frame3

animation.FuncAnimation(fig, animate, frames=opt.maxIter, interval=40, blit=True)
```
