```python
from SimPEG import *
from simpegem1d import EM1D, EM1DAnal, Kernels, RTEfun, DigFilter, BaseEM1D, Waveform
from scipy.constants import mu_0
from scipy.interpolate import interp1d
%pylab inline
```

```python
TDsurvey = EM1D.BaseEM1D.EM1DSurveyTD()
TDsurvey.rxLoc = np.array([0., 0., 100.+30.])
TDsurvey.srcLoc = np.array([0., 0., 100.+30.])
TDsurvey.fieldtype = 'secondary'
TDsurvey.rxType = 'Bz'
TDsurvey.waveType = 'stepoff'
TDsurvey.srcType = 'CircularLoop'
I = 1e0
a = 1e1
TDsurvey.I = I
TDsurvey.a = a

nearthick = np.logspace(-1, 1, 2)
deepthick = np.logspace(1, 2, 2)
hx = np.r_[nearthick, deepthick]
mesh1D = Mesh.TensorMesh([hx], [0.])
depth = -mesh1D.gridN[:-1]
LocSigZ = -mesh1D.gridCC
nlay = depth.size
topo = np.r_[0., 0., 100.]
TDsurvey.depth = depth
TDsurvey.topo = topo
TDsurvey.LocSigZ = LocSigZ
TDsurvey.HalfSwitch = True
TDsurvey.Setup1Dsystem()
# TDsurvey.time = np.logspace(-16, -2.5, 1024)
TDsurvey.time = np.logspace(-5, -2, 56)

TDsurvey.setFrequency(TDsurvey.time)
chi_half = 0.
```

```python
Expmap = BaseEM1D.BaseEM1DMap(mesh1D)

tau_blk = 0.0002
eta_blk = 0.5
c_blk = 1.0
# blkind = (LocSigZ>-200) & (LocSigZ<-100)
blkind = 0
sig_half = 0.001 
sig_blk = 0.001
sig = np.ones(nlay)*sig_half
sig[blkind] = sig_blk
tau = np.zeros(nlay)
tau[blkind] = tau_blk
eta = np.zeros(nlay)
eta[blkind] = eta_blk
c = np.zeros(nlay)
c[blkind] = c_blk
# options = {'Frequency': TDsurvey.frequency, 'tau': np.ones(nlay)*tau_half, 'eta':np.ones(nlay)*eta_half, 'c':np.ones(nlay)*c_half}
optionsCole = {'Frequency': TDsurvey.frequency, 'tau': tau, 'eta':eta, 'c':c}
modelReal = Expmap
modelComplex = BaseEM1D.BaseColeColeMap(mesh1D, **optionsCole)*Expmap           
m_1D = np.log(np.ones(nlay)*sig)

WT0, WT1, YBASE = DigFilter.LoadWeights()
options = {'WT0': WT0, 'WT1': WT1, 'YBASE': YBASE}
# Question to rowan: can I share same survey object
```

```python
tau_1 = np.zeros(nlay)
tau_1[blkind] = 0.0001
optionsCole = {'Frequency': TDsurvey.frequency, 'tau': tau_1, 'eta':eta, 'c':c}
modelComplex_1 = BaseEM1D.BaseColeColeMap(mesh1D, **optionsCole)*Expmap           
```

```python
probEM = EM1D.EM1D(mesh1D, modelReal, **options)
if probEM.ispaired:
    probEM.unpair()
if TDsurvey.ispaired:
    TDsurvey.unpair()
probEM.pair(TDsurvey)
probEM.chi = np.zeros(TDsurvey.nlay)
probEM.CondType = 'Real'        
m_1D = np.log(sig)
probEM.jacSwitch = True
Hz, dHzdsig = probEM.fields(m_1D)
probEM.survey.rxType = 'Bz'
BzTD = probEM.survey.projectFields(u=Hz)
dBzdsigTD = probEM.survey.projectFields(u=dHzdsig)
probEM.survey.rxType = 'dBzdt'
dBzdtTD = probEM.survey.projectFields(u=Hz)
ddBzdtdsigTD = probEM.survey.projectFields(u=dHzdsig)
```

```python
probEMIP = EM1D.EM1D(mesh1D, modelComplex, **options)
if probEMIP.ispaired:
    probEMIP.unpair()
if TDsurvey.ispaired:
    TDsurvey.unpair()
probEMIP.pair(TDsurvey)
probEMIP.chi = np.zeros(TDsurvey.nlay)
probEMIP.CondType = 'Complex'        
m_1D = np.log(sig)
probEMIP.survey.rxType = 'Bz'
HzCole = probEMIP.fields(m_1D)
BzTDCole = probEMIP.survey.projectFields(u=HzCole)
probEMIP.survey.rxType = 'dBzdt'
dBzdtTDCole = probEMIP.survey.projectFields(u=HzCole)
```

```python
probEMIP = EM1D.EM1D(mesh1D, modelComplex_1, **options)
if probEMIP.ispaired:
    probEMIP.unpair()
if TDsurvey.ispaired:
    TDsurvey.unpair()
probEMIP.pair(TDsurvey)
probEMIP.chi = np.zeros(TDsurvey.nlay)
probEMIP.CondType = 'Complex'        
m_1D = np.log(sig)
probEMIP.survey.rxType = 'Bz'
HzCole_1 = probEMIP.fields(m_1D)
BzTDCole_1 = probEMIP.survey.projectFields(u=HzCole_1)
probEMIP.survey.rxType = 'dBzdt'
```

```python
freqind = np.logical_and(TDsurvey.frequency>1e-1, TDsurvey.frequency<1e8)
frequency = TDsurvey.frequency[freqind]
BzEM = -Hz[freqind]*mu_0
BzEMIP = -HzCole[freqind]*mu_0
BzEMIP_1 = -HzCole_1[freqind]*mu_0
```

```python
BzIP = BzEMIP - BzEM
BzIP_1 = BzEMIP_1 - BzEM
```

```python
BzTDIP = BzTDCole - BzTD
BzTDIP_1 = BzTDCole_1 - BzTD
```
