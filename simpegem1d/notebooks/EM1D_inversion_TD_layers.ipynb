```python
from SimPEG import *
from simpegem1d import EM1D, EM1DAnal, Kernels, RTEfun, DigFilter, BaseEM1D
%pylab inline
```

```python
TDsurvey = BaseEM1D.EM1DSurveyTD()
TDsurvey.rxLoc = np.array([0., 0., 100.+30.])
TDsurvey.txLoc = np.array([0., 0., 100.+80.])
TDsurvey.fieldtype = 'secondary'
TDsurvey.rxType = 'dBzdt'
TDsurvey.waveType = 'stepoff'
TDsurvey.txType = 'VMD'
TDsurvey.offset = 1e-5
# TDsurvey.a = 2.5
# TDsurvey.I = 1.
cs = 10.
nearthick = np.logspace(-1, 1, 3)
linthick = np.ones(10)*cs
deepthick = np.logspace(1, 2, 3)
hx = np.r_[nearthick, linthick, deepthick, deepthick[-1]]
mesh1D = Mesh.TensorMesh([hx], [0.])
depth = -mesh1D.gridN[:-1]
LocSigZ = -mesh1D.gridCC
nlay = depth.size
topo = np.r_[0., 0., 100.]
TDsurvey.depth = depth
TDsurvey.topo = topo
TDsurvey.LocSigZ = LocSigZ
TDsurvey.time = np.logspace(-5, -3, 31)
TDsurvey.switchInterp = True
TDsurvey.setFrequency(TDsurvey.time)
TDsurvey.Setup1Dsystem()
sig_half = 1e-3
sig_blk = 1e-2
chi_half = 0.
Logmodel = BaseEM1D.BaseEM1DModel(mesh1D)
model = Model.ComboModel(mesh1D, [Logmodel])
```

```python
sig  = np.ones(nlay)*sig_half
blk_ind = (-30>LocSigZ) & (-70<LocSigZ)
sig[blk_ind] = sig_blk
m_true = np.log(sig)
```

```python
fig, ax = subplots(1,1, figsize=(5, 8))
ax.plot(sig, LocSigZ,  'b.-')
ax.grid(color='k', alpha=0.5, linestyle='dashed', linewidth=0.5)
ax.set_xscale('log')
ax.set_xlim(1e-4, 1e1)
```

```python
%%time
WT0, WT1, YBASE = DigFilter.LoadWeights()
options = {'WT0': WT0, 'WT1': WT1, 'YBASE': YBASE}
prob = EM1D.EM1D(model, **options)
if prob.ispaired:
    prob.unpair()
if TDsurvey.ispaired:
    TDsurvey.unpair()
prob.pair(TDsurvey)
prob.CondType = 'Real'
prob.chi = np.zeros(TDsurvey.nlay)
d_true = TDsurvey.dpred(m_true)
```

```python
fig, axes = subplots(1,1, figsize = (7,5))
axes.plot(TDsurvey.time, -d_true, '.-')
axes.set_xscale('log');
axes.set_yscale('log');
```

```python
TDsurvey.dtrue = d_true
std = 0.05
noise = std*abs(TDsurvey.dtrue)*np.random.randn(*TDsurvey.dtrue.shape)
TDsurvey.dobs = TDsurvey.dtrue+noise
TDsurvey.std = TDsurvey.dobs*0 + std
a  = 100.
weight = (1-exp(-a*TDsurvey.time))*np.linalg.norm(TDsurvey.dobs)
TDsurvey.Wd = 1/(abs(TDsurvey.dobs)*std)
# TDsurvey.Wd = 1/(weight)
prob.jacSwitch = True
```

```python
fig, axes = subplots(1,1, figsize = (7,5))
axes.plot(TDsurvey.time, -d_true, 'b.-')
axes.plot(TDsurvey.time, -TDsurvey.dobs, 'r.-')
axes.set_xscale('log');
axes.set_yscale('log');
```

```python
opt = Optimization.InexactGaussNewton(maxIter = 10)
reg = Regularization.Tikhonov(model)
obj = ObjFunction.BaseObjFunction(TDsurvey, reg, beta=beta)
# Create an inversion object
inv = Inversion.BaseInversion(obj, opt)
m0 = np.log(np.ones(TDsurvey.nlay)*sig_half)
reg.alpha_s = 1e-5
reg.alpha_x = 1.
C =  Utils.Counter()
prob.counter = C
opt.counter = C
opt.LSshorten = 0.5
opt.remember('xc')
beta = Parameters.BetaSchedule(coolingFactor = 8, coolingRate = 1, beta0_ratio = 1e2)
inv.objFunc.beta = beta
```

```python
mopt = inv.run(m0)
```

```python
opt.counter.summary()
xc = opt.recall('xc')
```

```python
%%time
dpredall = []
maxIter = len(xc)
prob.jacSwitch = False
for i in range(maxIter):
    dpredtemp = TDsurvey.dpred(xc[i])
    dpredall.append(dpredtemp)
```

```python
from JSAnimation import IPython_display
from matplotlib import animation

fig, ax = subplots(1,2, figsize=(14, 8))
ax[0].plot(sig, LocSigZ,  'b.-', lw=2)
ax[0].grid(color='k', alpha=0.5, linestyle='dashed', linewidth=0.5)
ax[1].grid(color='k', alpha=0.5, linestyle='dashed', linewidth=0.5)

ax[1].plot(TDsurvey.time, -(TDsurvey.dobs), 'r.-', lw=2)
ax[0].set_xscale('log')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[0].set_xlim(1e-5, 1e0)


frame1, = ax[0].plot([], [],  'r.-', lw=2)
frame2, = ax[1].plot([], [],  'b.-' , lw=2)

def animate(i_id):
    
    frame1.set_data(model.transform(xc[i_id])*np.ones(LocSigZ.size), LocSigZ)
    frame2.set_data(TDsurvey.time, -(dpredall[i_id]))
    
    ax[0].legend(('True', 'Pred'), loc=1, fontsize = 14)
    ax[1].legend(('True', 'Pred'), loc=1, fontsize = 14)
    
    ax[0].set_title('$\sigma$ model, Iter = ' + str(i_id+1), fontsize = 16)
    ax[1].set_title('Pred and Obs data, Iter = ' + str(i_id+1), fontsize = 16)
    return frame1, frame2

animation.FuncAnimation(fig, animate, frames=maxIter, interval=40, blit=True)
```
