```python
from SimPEG import *
from simpegem1d import EM1D, EM1DAnal, Kernels, RTEfun, DigFilter, BaseEM1D
%pylab inline
```

```python
TDsurvey = BaseEM1D.EM1DSurveyTD()
TDsurvey.rxLoc = np.array([0., 0., 100.+1e-5])
TDsurvey.txLoc = np.array([0., 0., 100.+1e-5])
TDsurvey.fieldtype = 'secondary'
TDsurvey.rxType = 'Bz'
TDsurvey.waveType = 'stepoff'
TDsurvey.txType = 'CircularLoop'
I = 1e0
a = 1e1
TDsurvey.I = I
TDsurvey.a = a

nearthick = np.logspace(-1, 1, 5)
deepthick = np.logspace(1, 2, 10)
hx = np.r_[nearthick, deepthick]
mesh1D = Mesh.TensorMesh([hx], [0.])
depth = -mesh1D.gridN
LocSigZ = np.r_[-mesh1D.gridCC, -mesh1D.gridCC[-1]-mesh1D.gridCC[-1]]
nlay = depth.size
topo = np.r_[0., 0., 100.]
TDsurvey.depth = depth
TDsurvey.topo = topo
TDsurvey.LocSigZ = LocSigZ
TDsurvey.Setup1Dsystem()
TDsurvey.time = np.logspace(-5, -2.5, 64)
TDsurvey.setFrequency(TDsurvey.time)
sig_half = 1e-4
sig_blk = 1e-2
chi_half = 0.

Logmodel = BaseEM1D.BaseEM1DModel(mesh1D)
tau_blk = 1e-3
eta_blk = 5e-1
c_blk = 0.8
blkind = (LocSigZ>-200) & (LocSigZ<-100)

sig = np.ones(nlay)*sig_half
sig[blkind] = sig_blk
tau = np.zeros(nlay)
tau[blkind] = tau_blk
eta = np.zeros(nlay)
eta[blkind] = eta_blk
c = np.zeros(nlay)
c[blkind] = c_blk


# options = {'Frequency': TDsurvey.frequency, 'tau': np.ones(nlay)*tau_half, 'eta':np.ones(nlay)*eta_half, 'c':np.ones(nlay)*c_half}
options = {'Frequency': TDsurvey.frequency, 'tau': tau, 'eta':eta, 'c':c}
Colemodel = BaseEM1D.BaseColeColeModel(mesh1D, **options)

modelReal = Model.ComboModel(mesh1D, [Logmodel])
modelComplex = Model.ComboModel(mesh1D, [Colemodel, Logmodel])                
m_1D = np.log(np.ones(nlay)*sig)

TDsurvey.rxType = 'Bz'        

WT0 = np.load('../WT0.npy')
WT1 = np.load('../WT1.npy')
YBASE = np.load('../YBASE.npy')
options = {'WT0': WT0, 'WT1': WT1, 'YBASE': YBASE}
# Question to rowan: can I share same survey object
```

```python
probEM = EM1D.EM1D(modelReal, **options)
if probEM.ispaired:
    probEM.unpair()
if TDsurvey.ispaired:
    TDsurvey.unpair()
probEM.pair(TDsurvey)
probEM.chi = np.zeros(TDsurvey.nlay)
print probEM.survey.nlay

probEM.CondType = 'Real'        
# probEM.survey.offset = 10.
sig_half = 0.01
m_1D = np.log(sig)
Hz = probEM.fields(m_1D)
probEM.survey.rxType = 'Bz'
BzTD = probEM.survey.projectFields(u=Hz)
probEM.survey.rxType = 'dBzdt'
dBzdtTD = probEM.survey.projectFields(u=Hz)
```

```python
probEMIP = EM1D.EM1D(modelComplex, **options)
if probEMIP.ispaired:
    probEMIP.unpair()
if TDsurvey.ispaired:
    TDsurvey.unpair()
probEMIP.pair(TDsurvey)
probEMIP.chi = np.zeros(TDsurvey.nlay)
probEMIP.CondType = 'Complex'        
# probEM.survey.offset = 10.
sig_half = 0.01
m_1D = np.log(sig)
probEMIP.survey.rxType = 'Bz'
HzCole = probEMIP.fields(m_1D)
BzTDCole = probEMIP.survey.projectFields(u=HzCole)
probEMIP.survey.rxType = 'dBzdt'
dBzdtTDCole = probEMIP.survey.projectFields(u=HzCole)
```

```python
fig, axes = subplots(1,2, figsize = (12,5))
axes[0].loglog(probEMIP.survey.time, abs(BzTD), 'b')
axes[0].loglog(probEMIP.survey.time, abs(BzTDCole), 'b*')
axes[1].loglog(probEMIP.survey.time, abs(dBzdtTD), 'r')
axes[1].loglog(probEMIP.survey.time, abs(dBzdtTDCole), 'r*')
```

```python
dbip = BzTDCole-BzTD; indb = dbip>0
dbdtip = -(dBzdtTDCole-dBzdtTD); inddbdt = dbdtip>0
fig, axes = subplots(1,2, figsize = (12,5))
axes[0].loglog(probEMIP.survey.time, abs(BzTD), 'k')
axes[0].loglog(probEMIP.survey.time[BzTDCole>0], BzTDCole[BzTDCole>0], 'b', probEMIP.survey.time[BzTDCole<0], -BzTDCole[BzTDCole<0], 'b--')
axes[0].loglog(probEMIP.survey.time[indb], dbip[indb], 'r')
axes[0].loglog(probEMIP.survey.time[~indb], -dbip[~indb], '--r')
axes[1].loglog(probEMIP.survey.time, abs(dBzdtTD), 'k')
axes[1].loglog(probEMIP.survey.time[-dBzdtTDCole>0], -dBzdtTDCole[-dBzdtTDCole>0], 'b', probEMIP.survey.time[-dBzdtTDCole<0], dBzdtTDCole[-dBzdtTDCole<0], 'b--')
axes[1].loglog(probEMIP.survey.time[inddbdt], dbdtip[inddbdt], 'r')
axes[1].loglog(probEMIP.survey.time[~inddbdt], -dbdtip[~inddbdt], '--r')
```
